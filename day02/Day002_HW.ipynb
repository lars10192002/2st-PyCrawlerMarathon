```python
# 2. 將班次一的每一個時間用一種資料型態保存

import csv

# 開啟 CSV 檔案
with open('./data/example.csv', newline='') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.DictReader(csvfile )#OrderedDict
    # get first row
    row = next(rows)
    print(row['\ufeff序號'],row['班次1'],row['班次2'],row['班次3'],row['班次4'],row['班次5'],row['班次6'],row['班次7'],row['班次8'],row['班次9'],row['班次10'],row['班次11'])
    print('--------------------------------------------------------------------')
    for k,v in row.items():
        print(k,v)

```

```python
# 2. 將班次一的每一個時間用一種資料型態保存 ANS

data = []
with open('./data/example.csv', newline='') as csvfile:
  # 讀取 CSV 檔案內容
  rows = csv.reader(csvfile)
  # 以迴圈輸出每一列
  for row in rows:
    data.append(row[5])
print(data)
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存

import csv

# 開啟 CSV 檔案
with open('./data/example.csv', newline='') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.DictReader(csvfile )#OrderedDict
    for row in rows:
        for k,v in row.items():
            print(k,v)
    
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存 ANS

d1 = []
d2 = []
d3 = []
d4 = []
d5 = []

with open('./data/example.csv', newline='') as csvfile:
  # 讀取 CSV 檔案內容
  rows = csv.reader(csvfile)
  # 以迴圈輸出每一列
  for row in rows:
    d1.append(row[5])
    d2.append(row[6])
    d3.append(row[7])
    d4.append(row[8])
    d5.append(row[9])

data = {}
data[d1[0]] = d1[1:]
data[d2[0]] = d2[1:]
data[d3[0]] = d3[1:]
data[d4[0]] = d4[1:]
data[d5[0]] = d5[1:]

data
```
