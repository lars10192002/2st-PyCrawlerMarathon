### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text,'html5lib')


a=soup.find("div",class_="c1").find("div",class_="part_list_2").find_all("h3")
em=soup.find("div",class_="c1").find("div",class_="part_list_2").find_all("em")

```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
res={}
for tag,title in zip(em,a):
    if(tag.text not in res):
        res.update({tag.text:[title.text]})
    else:
        res[tag.text].append(title.text)


res
```

### ③ 哪一個類別的文章最多

```python
tmp = 0
for i in res:
    print("{}:{}".format(i,len(res[i])))
    if(tmp<len(res[i])):
        tmp=len(res[i])
        k = i

print("最多是: "+k,"共: "+str(tmp)+"個")
```
